```python
#Import our modules that we are using
import matplotlib.pyplot as plt
import numpy as np

NdE=np.arange(10,10000000,100000);
NaE=1;
ni=1000000000;
NdI=np.arange(1,101);
NaI=1;
n0E=(NdE-NaE)/2 + (((NdE-NaE)/2)**(2)+(ni)**2)**(1/2);
n0I=(NdI-NaI)/2 + (((NdI-NaI)/2)**(2)+(ni)**2)**(1/2);
# Create the plot
plt.loglog(NdE,n0E,NdE,n0I)
plt.title('Concen. of Electrons vs Concen. of Donor Atoms')
plt.legend(['Extrinsic','Intrinsic'])
plt.xlabel('Donor Concentration')
plt.ylabel('Concentration of Electrons')

# Show the plot
plt.show()
```

```python
#Import our modules that we are using
import matplotlib.pyplot as plt
import numpy as np

NaE=10000000;
NdE=np.arange(10,10000000,100000);
ni=1000000000;
NaI=1000;
NdI=np.arange(1,101);
npE=(NaE-NdE)/2 + (((NaE-NdE)/2)**(2)+(ni)**2)**(1/2);
npI=(NaI-NdI)/2 + (((NaI-NdI)/2)**(2)+(ni)**2)**(1/2);
# Create the plot
plt.loglog(NdE,npE,NdE,npI)
plt.title('Concen. of holes vs Concen. of Donor Atoms')
plt.legend(['Extrinsic','Intrinsic'])
plt.xlabel('Donor Concentration')
plt.ylabel('Concentration of holes')

# Show the plot
plt.show()
```

```python
e=1.6*10**(-19);
E=100;
un=1350;
NdE=np.arange(10,100000,10000);
NdI=np.arange(1,11)
JdifE=e*un*NdE*E;
JdifI=e*un*NdI*E;
# Create the plot
plt.loglog(NdE,JdifE,NdE,JdifI)
plt.title('Drift Current vs Conc. of Donor Atoms')
plt.legend(['Extrinsic','Intrinsic'])
plt.xlabel('Donor Concentration')
plt.ylabel('Drift Current')

# Show the plot
plt.show()
```

```python


RE=E/JdifE;
RI=E/JdifI;

# Create the plot
plt.loglog(NdE,RE,NdE,RI)
plt.title('Resistivity vs Conc. of Donor Atoms')
plt.xlabel('Donor Concentration')
plt.ylabel('Resistivity')

# Show the plot
plt.show()
```
